```python
import pandas as pd
import matplotlib.pyplot as plt

# Define your columns here
COLUMN_A = 'sepal_length'
COLUMN_B = 'sepal_width'
CALORIES_COLUMN = 'sepal_length'  # Using 'sepal_length' for example
PULSE_COLUMN = 'sepal_width'      # Using 'sepal_width' for example
MAXPULSE_COLUMN = 'petal_length'  # Example column to drop
DURATION_COLUMN = 'petal_width'   # Example column for scatter plot

# Step 1: Read the CSV File
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
df = pd.read_csv(url, header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])

# Step 2: Show Basic Statistical Description
print("Basic Statistical Description:")
print(df.describe())

# Step 3: Check for Null Values
print("\nNull Values in Each Column:")
print(df.isnull().sum())

# Step 4: Replace Null Values with Mean
df.fillna(df.mean(numeric_only=True), inplace=True)

# Step 5: Aggregate Data for Two Columns
if COLUMN_A in df.columns and COLUMN_B in df.columns:
    print(f"\nAggregation for columns '{COLUMN_A}' and '{COLUMN_B}':")
    aggregation = df[[COLUMN_A, COLUMN_B]].agg(['min', 'max', 'count', 'mean'])
    print(aggregation)
else:
    print(f"\nColumns '{COLUMN_A}' and '{COLUMN_B}' do not exist in the DataFrame.")

# Step 6: Filter Rows with a Hypothetical 'Calories' Range
# Note: This dataset does not have a 'Calories' column, so this step will be skipped.
print("\nSkipping filter based on 'Calories' as it does not exist in this dataset.")

# Step 7: Filter Rows with Hypothetical Conditions on 'Calories' and 'Pulse'
# Note: This dataset does not have 'Calories' or 'Pulse' columns, so this step will be skipped.
print("\nSkipping filter based on 'Calories' and 'Pulse' as they do not exist in this dataset.")

# Step 8: Create a New DataFrame Without 'petal_length'
if MAXPULSE_COLUMN in df.columns:
    df_modified = df.drop(columns=[MAXPULSE_COLUMN])
    print(f"\nDataFrame Without '{MAXPULSE_COLUMN}':")
    print(df_modified)
else:
    print(f"\nColumn '{MAXPULSE_COLUMN}' does not exist in the DataFrame.")

# Step 9: Delete the 'petal_length' Column from the Original DataFrame
if MAXPULSE_COLUMN in df.columns:
    df.drop(columns=[MAXPULSE_COLUMN], inplace=True)
    print(f"\nOriginal DataFrame After Deleting '{MAXPULSE_COLUMN}':")
    print(df)
else:
    print(f"\nColumn '{MAXPULSE_COLUMN}' does not exist in the DataFrame.")

# Step 10: Convert Data Type of 'sepal_length' Column to Integer
if CALORIES_COLUMN in df.columns:
    df[CALORIES_COLUMN] = df[CALORIES_COLUMN].astype(int)
    print("\nData Types After Converting 'sepal_length' to Integer:")
    print(df.dtypes)
else:
    print(f"\nColumn '{CALORIES_COLUMN}' does not exist in the DataFrame.")

# Step 11: Create a Scatter Plot for 'petal_width' and 'sepal_length'
if DURATION_COLUMN in df.columns and CALORIES_COLUMN in df.columns:
    plt.scatter(df[DURATION_COLUMN], df[CALORIES_COLUMN])
    plt.xlabel('Petal Width')
    plt.ylabel('Sepal Length')
    plt.title('Scatter Plot of Petal Width vs. Sepal Length')
    plt.show()
else:
    print(f"\nOne or both columns '{DURATION_COLUMN}' and '{CALORIES_COLUMN}' do not exist in the DataFrame.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# a) Import the given “Iris” dataset CSV
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
# The Iris dataset does not have a column for salary. Let's assume we use SepalLength and PetalLength for this demo.
df = pd.read_csv(url, header=None, names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Class'])

# Display the first few rows of the dataframe to understand its structure
print("First few rows of the dataset:")
print(df.head())

# Select features and target variable
# For demonstration, we use 'SepalLength' as the feature and 'PetalLength' as the target
X = df[['SepalLength']]
y = df['PetalLength']

# b) Split the data into train and test partitions
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

# c) Train and predict the model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on both train and test data
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

# d) Calculate the mean squared error
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)

print(f"Mean Squared Error on Training Data: {mse_train}")
print(f"Mean Squared Error on Testing Data: {mse_test}")

# e) Visualize both train and test data using scatter plot
plt.figure(figsize=(14, 6))

# Plot Training Data
plt.subplot(1, 2, 1)
plt.scatter(X_train, y_train, color='blue', label='Training data')
plt.plot(X_train, y_train_pred, color='red', linewidth=2, label='Fitted line')
plt.title('Training Data')
plt.xlabel('SepalLength')
plt.ylabel('PetalLength')
plt.legend()

# Plot Testing Data
plt.subplot(1, 2, 2)
plt.scatter(X_test, y_test, color='green', label='Testing data')
plt.plot(X_train, model.predict(X_train), color='red', linewidth=2, label='Fitted line')
plt.title('Testing Data')
plt.xlabel('SepalLength')
plt.ylabel('PetalLength')
plt.legend()

plt.tight_layout()
plt.show()
```
